```python
from pandas_datareader import data as wb
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = wb.DataReader('SPY','yahoo', '2015-7-6')
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA','Close']].plot(figsize=(14,6))
```

```python
#Considers all the averages from the very begining.
df['Close'].expanding().mean().plot(figsize=(14,6))
```

**Bollinger Bands**

```python
#Close 20 MA
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()

#Upper = 20MA + 2 * STD(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())

#Lower = 20MA + 2 * STD(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())

#Close
df[['Close', 'Close: 20 Day Mean', 'Upper','Lower']].plot(figsize=(16,7))
```
