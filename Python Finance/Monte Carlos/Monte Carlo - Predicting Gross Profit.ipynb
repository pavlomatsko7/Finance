```python
import numpy as np
import matplotlib.pyplot as plt

rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
#Assuming COGS is around 60% of revenues and deviates 10%
COGS = -(rev * np.random.normal(0.6,0.1))

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
print('COGS mean:',str(COGS.mean()))
print('COGS stdev:',str(COGS.std()))
```

```python
#Used a minus sign above, so now a plus
Gross_Profit = rev + COGS
Gross_Profit

plt.figure(figsize=(15,6))
plt.plot(Gross_Profit)
plt.show()
```

```python
print('Max Gross Profit:', str(max(Gross_Profit)))
print('Min Gross Profit:', str(min(Gross_Profit)))
```

```python
print('Gross Profit mean:', str(Gross_Profit.mean()))
print('Gross Profit stdev:', str(Gross_Profit.std()))
```

```python
#Bins grouping inbetween 
plt.figure(figsize=(10,6));
plt.hist(Gross_Profit, bins= [40,50,60,70,80,90,100,110,120]);
plt.show()
```
