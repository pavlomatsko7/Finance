```python
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
spy = wb.DataReader('SPY', data_source='yahoo', start='1993-1-1')
```

**Simple Rate of Return**
$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
#Calculation
spy['simple_return'] = (spy['Adj Close'] / spy['Adj Close'].shift(1)) - 1
#Daily % Returns
spy['simple_return'].plot(figsize=(8,5))
plt.show()
```

```python
avg_return_d = spy['simple_return'].mean()
print("Average daily % return:",str(round(avg_return_d, 7) * 100) + '%')
avg_return_a = spy['simple_return'].mean() * 252
print('Average yearly % return:', str(round(avg_return_a, 4) * 100) + '%')
```

**Log Returns**
$$
ln(\frac{P_t}{P_{t-1}})
$$

```python
spy['log_return'] = np.log(spy['Adj Close'] / spy['Adj Close'].shift(1))
spy['log_return'].plot(figsize=(8,5))
plt.show()
```

```python
log_avg_return_d = spy['log_return'].mean()
print("Average daily log % return:",str(round(log_avg_return_d, 7) * 100) + '%')
log_avg_return_a = spy['log_return'].mean() * 252
print('Average yearly log % return:', str(round(log_avg_return_a, 4) * 100) + '%')
```
